## What is Functional Programming in Python?

## Real-World Example: Employee Salary Processing

A company has thousands of employee salaries. Instead of using multiple for loops, functional programming helps process the data in a concise and efficient way.

```python
from functools import reduce

salaries = [25000, 35000, 45000, 55000, 65000]

# Increase salary by 10%
updated_salary = list(map(lambda x: x * 1.10, salaries))

# Employees earning more than ₹40,000
high_salary = list(filter(lambda x: x > 40000, updated_salary))

# Total salary payout
total_salary = reduce(lambda x, y: x + y, high_salary)

print(updated_salary)
print(high_salary)
print(total_salary)
```

## WAP to print number is prime or not by using FP ( Functional programming )



```python
def prime():
  num = int(input("enter the number :"))
  if num > 1:
    for i in range(2,num):
      if num % i == 0:
        print(f"{num} is not prime number")
        break
    else:
      print(f"{num} is prime number")
prime()
```

 In Python, when you define a function with parameters followed by a slash (/),

 like def Addition(num1, num2, /):, it means that num1 and num2 can only be passed by their position,

  not by using their names as keywords. The second call, Addition(num1 = 50, num2 = 70), attempts to pass these arguments as keyword arguments,

```python

def Addition(num1, num2,/): # parameters
    add = num1 + num2
    print(add)

Addition(10,20) # arguments
Addition(num1 = 50, num2 = 70) # TypeError: Addition() got some positional-only arguments passed as keyword arguments: 'num1, num2'
```

## List Comprehension

A list comprehension is a concise way to create a list using a single line of code.

```python
numbers = [1, 2, 3, 4, 5]

squares = [x * x for x in numbers]

print(squares)
```

## Dictionary Comprehension

A concise way to create a dictionary by generating key-value pairs from an iterable.

```python
numbers = [1, 2, 3, 4, 5]

square_dict = {x: x * x for x in numbers}

print(square_dict)
```

## Set Comprehension

A concise way to create a set by generating unique values from an iterable.

```python
numbers = [1, 2, 2, 3, 4, 4, 5]

unique_numbers = {x for x in numbers}

print(unique_numbers)
```

## Generator Expression

A concise way to create a generator that produces values one at a time, making it memory efficient.

```python
numbers = [1, 2, 3, 4, 5]

squares = (x * x for x in numbers)

for value in squares:
    print(value)
```
