## II ) Variable

In python variable is like a container which is used to store a diffrent types of datatypes, A variable is a named reference that points to an object stored in memory,no explicit data type declaration is required because Python is dynamically typed.

## syntax

var_name = Var_value


1. A variable is a named memory location used to store data.
2. Python is dynamically typed, so you don't need to declare the data type explicitly.
3. Variable names are case-sensitive (age and Age are different variables).
4. A variable can store any datatype, such as int, float, str, bool, list, tuple, dict, or custom objects.
5. Variables can be reassigned to values of different data types.
6. Python manages memory automatically using garbage collection.
7. Multiple variables can reference the same object.

Certainly! The code defines a function called is_leap that checks if a given year is a leap year. Let me break down how it works:


leap = False: It initializes a boolean variable leap to False. This variable will store the result.


if year % 4 == 0:: It first checks if the year is divisible by 4. If a year is not divisible by 4, it cannot be a leap year, so leap remains False.


if year % 100 == 0:: If the year is divisible by 4, it then checks if it's also divisible by 100. This is because years divisible by 100 are generally not leap years, unless...


if year % 400 == 0:: ...if a year divisible by 100 is also divisible by 400, then it is a leap year. In this case, leap is set to True.


else: leap = False: If the year is divisible by 100 but not by 400, then it's not a leap year, so leap is set to False.


else: leap = True: Going back to the year % 100 == 0 check, if the year is divisible by 4 but not by 100, then it is a leap year, and leap is set to True.


else: leap = False: This is the outermost else block. If the year is not divisible by 4 at all, then leap remains False.


Finally, the function returns the value of leap. The last two lines of the code year = int(input()) and print(is_leap(year)) prompt the user to enter a year, convert it to an integer, and then print whether it's a leap year or not by calling the is_leap function.

```python

def is_leap(year):
    leap = False
    if year %4 == 0:
        if year %100 == 0:
            if year %400 == 0:
                leap = True
            else:
                leap = False
        else:
            leap = True
    else:
        leap = False

    return leap

year = int(input())
print(is_leap(year))
```
