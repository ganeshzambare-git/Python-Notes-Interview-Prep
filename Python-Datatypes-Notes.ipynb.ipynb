## Variable

In python variable is like a container which is used to store a diffrent types of datatypes.

syntax

var_name = Var_value 

1. A variable is a named memory location used to store data.
2. Python is dynamically typed, so you don't need to declare the data type explicitly.
3. Variable names are case-sensitive (age and Age are different variables).
4. A variable can store any datatype, such as int, float, str, bool, list, tuple, dict, or custom objects.

## variable naming rules

1. variable names  can contain letters,numbers & underscore.
2. variable name must start with letter or underscore.
3. variable names are case sensitive.
4. avoid using python keyword as variable names (eg. if,else,print,append,class,while... etc.)
5. avoid using special caracters(ie @,#,<,%,>,) and dont start the var name with special char.

## Methods Applicable on Sets (set)

| Method                          | Purpose                                                                               |
| ------------------------------- | ------------------------------------------------------------------------------------- |
| `add()`                         | Adds a single element to the set.                                                     |
| `update()`                      | Adds multiple elements from another iterable.                                         |
| `remove()`                      | Removes a specified element. Raises an error if the element is not found.             |
| `discard()`                     | Removes a specified element. Does **not** raise an error if the element is not found. |
| `pop()`                         | Removes and returns a random element from the set.                                    |
| `clear()`                       | Removes all elements from the set.                                                    |
| `copy()`                        | Returns a shallow copy of the set.                                                    |
| `union()`                       | Returns a new set containing all unique elements from both sets.                      |
| `intersection()`                | Returns common elements between two or more sets.                                     |
| `difference()`                  | Returns elements present in the first set but not in the second.                      |
| `symmetric_difference()`        | Returns elements present in either set but not in both.                               |
| `intersection_update()`         | Updates the original set with common elements only.                                   |
| `difference_update()`           | Removes elements found in another set from the original set.                          |
| `symmetric_difference_update()` | Updates the original set with symmetric difference.                                   |
| `issubset()`                    | Checks whether one set is a subset of another.                                        |
| `issuperset()`                  | Checks whether one set is a superset of another.                                      |
| `isdisjoint()`                  | Checks whether two sets have no common elements.                                      |
